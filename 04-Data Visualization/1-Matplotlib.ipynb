# Matplotlib

Matplotlib is the library of data visualization with Python.

It is an excellent 2D and 3D graphics library for generating scientific figures. 

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

Matplotlib allows you to create reproducible figures programmatically.

### Example

Let's walk through a very simple example using numpy arrays. You can also use lists, but most likely you'll be passing numpy arrays or pandas columns (which essentially also behave like arrays).

** The data we want to plot:**

```python
import numpy as np
x = np.arange(20)
y1 = x
y2 = x ** 2
y3 = x ** 3
```

```python
x
```

```python
y1,y2,y3
```

## Basic Matplotlib Commands

We can create a very simple line plot using the following.

```python
plt.plot(x, y1, 'b',label="Linear") # 'b' is the color blue
plt.plot(x, y2, 'r',label="Square") # 'r' is the color red
plt.plot(x, y3, 'g',label="Cube") # 'g' is the color green
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Title')
plt.legend()
plt.show()
```

## Saving figures
Matplotlib can generate high-quality output in a number formats, including PNG, JPG, EPS, SVG, PGF and PDF. 

To save a figure to a file we can use the `savefig` method in the `Figure` class:

```python
plt.plot(x, y1, 'b',label="Linear") # 'b' is the color blue
plt.plot(x, y2, 'r',label="Square") # 'r' is the color red
plt.plot(x, y3, 'g',label="Cube") # 'g' is the color green
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Title')
plt.legend()
plt.savefig("filename.png")
```

## Plot range

```python
plt.plot(x, x**2, x, x**3)
plt.ylim([0, 60])
plt.xlim([2, 5])
plt.title("custom axes range");
```

```python
import numpy as np
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), 'b-')
plt.plot(x, np.cos(x), 'r--');
```

# Special Plot Types

There are many specialized plots we can create, such as barplots, histograms, scatter plots, and much more. Most of these type of plots we will actually create using pandas. But here are a few examples of these type of plots:

## Scatter Plot

```python
plt.scatter(x,y3)
```

## Histograms

```python
import numpy as np
plt.hist(np.random.randn(100))
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Title')
plt.legend()
plt.show()
```

## Bar Graph

```python
plt.bar([1,2,3],[15,15,35],.1)
plt.bar([1,2,3],[10,3,30],.1)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Title')
plt.legend()
plt.show()
```

```python
plt.bar(np.arange(10),(np.random.rand(10)*5)+4,.8)
plt.bar(np.arange(10),(np.random.rand(10)*5),.8)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Title')
plt.legend()
plt.show()
```

```python
w=4
n=10
s=5
R=np.arange(0,s*n*w,w*s)
S=np.arange(w,(s*n+1)*w,w*s)
plt.bar(R,(np.random.rand(n)*5)+4,w)
plt.bar(S,(np.random.rand(n)*5),w)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Title')
plt.legend()
plt.show()
```
