## Clustering
* It is the task of grouping a set of objects (usually unlabelled data) in such a way that objects in the same group (called a cluster) are more similar (in some sense) to each other than to those in other groups (clusters)






```python
import pandas as pd
from sklearn.cluster import KMeans, SpectralClustering, DBSCAN
from sklearn.datasets import make_moons, make_blobs
%matplotlib inline
```

The DBSCAN algorithm views clusters as areas of high density separated by areas of low density.

```python
X,y=make_blobs(n_samples=1000)
```

```python
# X,y=make_moons(n_samples=1000,noise=0.08, random_state=0)
```

```python
X=pd.DataFrame(data=X, columns=['x','y'])
X.head()
```

```python
X.plot.scatter(x='x',y='y')
```

```python
clf = SpectralClustering(n_clusters=3).fit(X)
```

```python
clf.labels_
```

```python
X.plot.scatter(x='x',y='y',c=clf.labels_,colormap='coolwarm')
```
