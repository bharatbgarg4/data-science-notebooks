“Ridge” is a fancy name for L2-regularization, “LASSO” means L1-regularization, “ElasticNet” is a ratio of L1 and L2 regularization. 

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn import metrics, linear_model as lm
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
data=pd.read_csv('../datasets/delta_elevators.csv')
data.head()
```

```python
X=data.iloc [:,:5].values
y=data.iloc [:,-1].values
X_train, X_test, y_train, y_test = train_test_split (X,y, test_size = .20) 
```

```python
clf= lm.Lasso()
clf = clf.fit(X_train, y_train)
```

```python
metrics.mean_absolute_error(y_test, clf.predict(X_test))
```

```python
data['Se'].iloc[:20].plot.line(figsize=(20,6))
```
