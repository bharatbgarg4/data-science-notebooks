# Machine Learning

Machine learning is a data analytics technique that teaches computers to do what comes naturally to humans and
animals: learn from experience. Machine learning algorithms use computational methods to “learn” information
directly from data without relying on a predetermined equation as a model. The algorithms adaptively improve their
performance as the number of samples available for learning increases.

Machine learning algorithms find natural patterns in data that generate insight and help you make
better decisions and predictions.

* computer programs that can access data and use it learn for themselves  
* the ability to automatically learn and improve from experience without being explicitly programmed  
*  look for patterns in data and make better decisions in the future based on the examples that we provide  

In Contrast,  
Programming is telling a computer what to do with the given set of instructions. It can range from making basic calculations such as multiplications to Apollo 11 instructions used in first moon landing. 

Both have different applications, like we use basic programming to make simple applications with repetative logic, but when it comes to make such programs where user interest needs to be taken care of, or study of a varying parameter is necessary, we use machine learning

## Supervised learning
* #### Availability of labelled data to train machine learning algorithm.
* #### By analysis this known training dataset, the learning algorithm produces an inferred function to make predictions about the output values.
* #### This function is later used to predice values in future.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
data = pd.read_csv ('datasets/Price.csv')
data.head()
```

```python
X = data.iloc [:,0].values
y = data.iloc [:,1].values
x1=X.copy()
X=X.reshape(-1, 1)
plt.scatter(X, y, color = 'green')
plt.title ('Random Data')
plt.xlabel('Area of Flat')
plt.ylabel('Price')
plt.show()
```

```python
x1, y1 = [2000, 1400], [90000, 70000]
plt.plot(x1, y1, marker = 'x')
plt.scatter(X, y, color = 'green')
plt.title ('Pattern Estimation')
plt.xlabel('Area of Flat')
plt.ylabel('Price')
plt.show()
```

## Classification:  
* #### samples belong to two or more classes   
* #### learn from already labeled data how to predict the class of unlabeled data

### Examples
* Cats vs Dogs
* Apples vs Oranges

```python
from sklearn.datasets import make_classification
X1, Y1 = make_classification(n_features=2, n_redundant=0, n_informative=2,
                             n_clusters_per_class=1, n_classes=3)
plt.scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1,
            s=25, edgecolor='k')
plt.title("Multi-class, two informative features, one cluster",
          fontsize='small')
plt.show()
```

```python
from sklearn.datasets import make_blobs
plt.title("Three blobs", fontsize='small')
X1, Y1 = make_blobs(n_features=2, centers=3)
plt.scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1,
            s=25, edgecolor='k')
plt.show()
```
