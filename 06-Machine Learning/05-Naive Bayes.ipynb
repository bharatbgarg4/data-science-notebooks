## Naive Bayes

* Naive Bayes treats each features as independent, whereas SVM looks at the interactions between them.

* SVM is better for features that have some form of relation, Naive Bayes performs better for independed features.

* Naive Bayes is faster than SVM, and works significantly better with large number of features.

* Naive Bayes works on Mathematical function based on Probability, whereas SVM works with Geomatric functions.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
import seaborn as sns
```

```python
data=pd.read_csv('../datasets/delta_elevators_class.csv')
data.head()
```

```python
sns.pairplot(data)
```

```python
X=data.iloc [:,:5].values
y=data.iloc [:,-1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split (X,y, test_size = .20) 
clf= GaussianNB()
# clf= SVC(kernel='rbf')
clf = clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
from sklearn import metrics
print(metrics.accuracy_score(y_test,y_pred))
```
